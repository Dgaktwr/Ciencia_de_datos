```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles['Moneda'] = valor[0]
inmuebles['Precio'] = valor[1]
inmuebles.sample(3)
```

```python
inmuebles.info()
```

```python
inmuebles['Precio'].sum()
```

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','',regex=True)
```

```python
inmuebles[['Precio','Barrio']]
```

```python
inmuebles.info()
```

```python
inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000
inmuebles.info()
```

```python
inmuebles.describe()
```

```python
pd.set_option('display.precision',2)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
inmuebles.describe()
```

```python
#importacion de graficas
import matplotlib.pyplot as plt
import seaborn as sns
#lienzo de la figura 10 ancho 8 el alto
plt.figure(figsize=(10,6))
#kde---> curva de densidad
#hue---> añade mas parametros en este caso el Tipo
grafica = sns.histplot(data=inmuebles, x='Precio_Millon', kde=True, hue='Tipo')
#titulo
grafica.set_title('Distribucion de valores de los inmubeles en Bogota')
#limitacion max de la variable
plt.xlim((50,1000))
#--->limitacion en y plt.ylim((0,20))
#guardar figura
plt.savefig('/content/drive/MyDrive/Ciencia de datos/valor_inmuebles.png',format='png')
plt.show()
```
